```python
hsv = cv2.cvtColor(np.float32(img),cv2.COLOR_BGR2HSV)
hue, sat, val = hsv[:,:,0], hsv[:,:,1], hsv[:,:,2]
h = np.mean(hsv[:,:,0])
s = np.mean(hsv[:,:,1])*100
v = ((np.mean(val))*100)/255
b = np.mean(img[:,:,0])
g = np.mean(img[:,:,1])
r = np.mean(img[:,:,2])
#cv2.imshow('HSV format image',hsv)
```
