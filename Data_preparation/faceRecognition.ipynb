```python
import cv2
import numpy as np
from PIL import Image 
import requests
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
url = 'https://scontent-lga3-1.cdninstagram.com/vp/1496a401b8ffa7b0765db8d19163371a/5E202CD9/t51.2885-15/e35/69770314_140073390581106_6781997854456604723_n.jpg?_nc_ht=scontent-lga3-1.cdninstagram.com&_nc_cat=106'
image = Image.open(requests.get(url, stream=True).raw)

pil_image = image.convert('RGB') 
image = np.array(pil_image) 
# Convert RGB to BGR 
image = image[:, :, ::-1].copy() 
 
# perform the actual resizing of the image and show it
image = cv2.resize(image, (500, 500), interpolation = cv2.INTER_AREA)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#type(image)
```

```python
faces = face_cascade.detectMultiScale(gray)
print (type(faces))
print (faces)
```

```python
print (faces.shape)
print ("number of faces detected: "+ str(faces.shape[0]))
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w, y+h), (0,255,0),1)
```

```python
cv2.rectangle(image, ((0, image.shape[0]-25)), (270, image.shape[0]), (255, 255, 255), -1)
cv2.putText(image, "number of faces detected: "+ str(faces.shape[0]), (0, image.shape[0]-10), cv2.FONT_HERSHEY_TRIPLEX, 0.5, (0, 0, 0), 1)
```

```python
cv2.imshow("image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
